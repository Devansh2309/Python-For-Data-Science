**PROJECT 3:- COVID DATA ANALYSIS**

```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/Project - 3(Covid Data).csv')
```

Q1:- Show the no. of confirmed cases, Deaths & Recovered Cases in each region?

```python
data.head(3)
data.groupby('Region')['Confirmed', 'Deaths', 'Recovered'].sum()
```

Q2:- Remove all the records where Confirmed Cases is Less than 10.

```python
data.head(2)
data[~(data.Confirmed<10)]
data
```

Q3:- In which region, maximum number of confirmed cases were recorded?

```python
data.head(2)
data.groupby('Region').Confirmed.sum().sort_values(ascending=False)
```

Q4:- In which region, maximum number of Deaths cases were recorded?

```python
data.head(2)
data.groupby('Region').Deaths.sum().sort_values()
```

Q5:- How many Confirmed, Deaths and Recovered cases were reported from India till 29th April 2020?

```python
data.head(10)
data[data.Region=='India']
```
