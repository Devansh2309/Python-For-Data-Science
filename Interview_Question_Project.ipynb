Q4:- Create a new column in the dataframe "age-group". The age groups are 18-25, 25-40, 40-55 & 55+.

```python
data2 = survey[(survey["Age_Group"]>=25) & (survey["Age_Group"]<40)]
data2['Age_Group'] = "25-40"
```

```python
survey = pd.concat([data1,data2,data3,data4])
survey
```
